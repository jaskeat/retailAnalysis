```python
#Import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
retailDf = pd.read_csv('../Data/SampleSuperstore.csv')
retailDf.head()
```

```python
retailDf.describe()
```

```python
retailDf.dtypes
```

# Data analysis

```python
# Calculate the distribution of categories
category_distribution = retailDf['Category'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
category_distribution.plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette())
plt.title('Category Distribution')
plt.ylabel('')  # Hide the y-label
plt.show()
```

```python
# Calculate the distribution of sub-categories
sub_category_distribution = retailDf['Sub-Category'].value_counts().head(10)

# Plot the bar chart
plt.figure(figsize=(12, 6))
sub_category_distribution.plot(kind='bar', color=sns.color_palette('muted'))
plt.title('Top 10 Sub-Categories')
plt.xlabel('Sub-Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by 'Category' and sum the 'Profit'
category_profit = retailDf.groupby('Category')['Profit'].sum()

# Plot the bar chart
plt.figure(figsize=(10, 6))
category_profit.plot(kind='bar', color=sns.color_palette('muted'))
plt.title('Total Profit by Category')
plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.xticks(rotation=0)
plt.show()
```
